```python
# This code allows you to try out geometric mappings and display them on a graph.
# - Try your own original shape instead of Duck.
# - Try your own geometric mappings to replace G1.
# - Try a sequence of geometric mappings.

# To use this, you will need to install matplotlib. On your command line, either:
#   pip install matplotlib
#   python3 -m pip install matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Add a row of ones to the end of a matrix of coordinates.
def add_homogeneous_row(M):
  return np.concatenate( (M, np.ones( (1,M.shape[1])) ))

# Plot the original shape with its transformation.
def plot_mapping(Original, Transformed):
  plt.plot(Original[0,:], Original[1,:], color='black', label = 'Original')
  plt.plot(Transformed[0,:], Transformed[1,:], color='black', linestyle='dashed', label = 'Transformed')
  plt.xlabel('x1')
  plt.ylabel('x2')
  plt.title('Geometric Mapping')
  plt.legend(loc = 'lower left')
  plt.show()
```

```python
# Matrix of coordinates for a duck
Duck = np.array([[2,1,2,3,5,6,5,3.5,3,2,1.8,1,  1.8,2], 
                 [2,1,0,0,0,1,2,2,  4,4,3.2,2.5,2.5,2]])
Duck = add_homogeneous_row(Duck)
print("This is the shape:")
print(Duck)

# Geometric mapping in 3x3 matrix: reflection in x1-axis & tranlation
G1 = np.array([[1,0,4], [0,-1,0], [0,0,1]])
print("This is the geometric mapping G1:")
print(G1)

# Alternative: horizontal shear & no translation
G2 = np.array([[1,2,0], [0,1,0], [0,0,1]])

# Compute the mapping and plot.
# Notice that "np.matmul" applies the geometric mapping to the shape.
plot_mapping(Duck, np.matmul(G1, Duck))
```
